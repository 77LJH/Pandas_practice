

Pandas mean()median()  mode() 

```python
#  mean() 
import pandas as pd

df=pd.read_csv("property-data.csv")

x=df["ST_NUM"].mean()

df["ST_NUM"].fillna(x,inplace=True)

print(df.to_string())
```

```python
#  median() 
import pandas as pd
df=pd.read_csv("property-data.csv")

x=df["ST_NUM"].median()

df["ST_NUM"].fillna(x,inplace=True)
print(df.to_string())
```

```python
#  mode() 
import pandas as pd
df=pd.read_csv("property-data.csv")

x=df["ST_NUM"].mode()

df["ST_NUM"].fillna(x,inplace=True)
print(df.to_string())
```

##  Pandas 

```python
import pandas as pd

person = {
  "name": ['Google', 'Runoob' , 'Taobao'],
  "age": [50, 40, 12345]    # 12345 
}

df=pd.DataFrame(person)

df.loc[2,"age"]=30 # 

print(df.to_string())
```

```python
#  age  120  120:
import pandas as pd

person = {
  "name": ['Google', 'Runoob' , 'Taobao'],
  "age": [50, 200, 12345]    
}

df = pd.DataFrame(person)

for x in df.index:
    if df.loc[x,"age"]>120:
        df.loc[x,"age"]=120

print(df.to_string())
```

```python
#  age  120 :
import pandas as pd

person = {
  "name": ['Google', 'Runoob' , 'Taobao'],
  "age": [50, 40, 12345]    # 12345 
}

df = pd.DataFrame(person)

for x in df.index:
    if df.loc[x,"age"]>120:
        df.drop(x,inplace=True)

print(df.to_string())
```

## Pandas 

 duplicated()  drop_duplicates() 

duplicated()  True False

```python
import pandas as pd

person = {
  "name": ['Google', 'Runoob', 'Runoob', 'Taobao'],
  "age": [50, 40, 40, 23]  
}
df=pd.DataFrame(person)

print(df.duplicated())
```

drop_duplicates() 

```python
import pandas as pd

persons = {
  "name": ['Google', 'Runoob', 'Runoob', 'Taobao'],
  "age": [50, 40, 40, 23]  
}

df = pd.DataFrame(persons)

df.drop_duplicates(inplace=True)

print(df)
```
